### Oppgave 1

En ratelov er produktet av pseudoratekonstanten og konsentrasjonene til reaktantene i reasjonen. En ratelov er forskjellig fra reaksjon til reaksjon, ut fra hvilke stoffer som reagerer med hverandre.

Rateloven er endring av produktet delt på tid.

Ratekoeffisienten bestemmes via Beers lov. Der bruker du observasjoner fra et forsøk til å finne konstanten til rateloven.
De andre faktorene bestemmes eksprimentelt, med tilnærmet valgfrie konsentrasjoner.


### Oppgave 2

Første eller andre ordens rate lov bestemmer vi ut i fra eksperimentet, eksperimentelt, og setter enten den ene, den andre, eller begge konsentrasjonene opphøyd i annen. Målet er å lage en modell som passer best mulig med svarene i eksperimentet.


Første ordens:

$v = c’(t) = k_r[CV^+][OH^-]$

Andre ordens:

$v = c’(t) = k_r[CV^+]^2[OH^-]^2$

---
$[OH^-]$ settes i nullte orden fordi vi i vårt forsøk hadde mye mer $OH^-$ enn $CV^+$. Dette gjør at konsentrasjonen av $OH^-$ er tilnærmet uforandret i forhold til $CV^+$. 
$[CV^+]$ setter vi i første orden fordi det er det som gir en lineær graf når vi tar logaritmen av $[CV^+]$. 


ln($[CV^+]$) = k’t + ln($[CV^+]_0$)

---
Det gjør at formelen for rateloven vil se slik ut:

Δ[CVOH]/Δt = k'$[CV^+]^1$$[OH^-]^0$

Δ[CVOH]/Δt = k'$[CV^+]^1$

### Oppgave 5

Bestemmer ratekoeffisienten via Beers lov

Beers lov:
$A = k_s*l*c$
---
Vi finner $k_s$ ved å sette inn :
- A (absorbsjonen av $CV^+$ ved start)
- l (bredden på kyvetten)
- c (konsentrasjonen av $CV^+$ ved start)


c vet vi allerede er $1.5*10^-5$ M og A finner vi som den første målingen ved forsøket

```python
A = 0.319
l = 1
c = 1.5E-5

k_s = A/(l*c)

print(k_s)
```

Når vi har funnet $k_s$ kan vi regne ut de konsentrasjonene som vi trenger for å finne k'. 
Med absorbsjonen regner vi ut konsentrasjonene av $[CV^+]$ ved start. Vi regner ut [CVOH] ved stat og [CVOH] ved slutt for å finne Δ[CVOH], ut fra $[CV^+]$ ved start.


```python
A_CV_start = 0.319   #Abrobsjon av CV ved start, funnet eksprimentelt
A_CV_slutt = 0.003
CVOH_start = A_CV_start/(k_s)  
CVOH_slutt = A_CV_slutt/(k_s)

delta_CVOH = CVOH_start-CVOH_slutt

k = (delta_CVOH)/(2500*1.41E-7)

print('k = ',k)
```

k' og k har samme verdi fordi:

$k' = k * [OH^-]^0$

og $[OH^-]^0$ er lik 1


$k' = k * 1$ 

# $k' = k$

### Oppgave 6

```python
CV_0 = 1.5E-5     #konsentrason av CV ved start
OH_0 = 0.05    #konsentrasjon av OH ved start
CVOH_0 = 0   #konsentrasjon av CVOH ved start


#Tidssteg
N = 100000 #antall intervller
tid = 2500    #tid i minutter
dt = tid/(N-1)

#Matriser
t = pl.zeros(N)
CV = pl.zeros(N)
OH = pl.zeros(N)
CVOH = pl.zeros(N)
CV_der = pl.zeros(N)
OH_der = pl.zeros(N)
CVOH_der = pl.zeros(N)


#Initierer matriser
CV[0] = CV_0
OH[0] = OH_0
CVOH[0] = CVOH_0



```

```python
#Eulers metode
for i in range(N-1):
    CVOH_der[i] = k*CV[i]*OH[i]
    CV_der[i] = - CVOH_der[i]
    OH_der[i] = CV_der[i]
    CV[i+1] = CV[i]+CV_der[i]*dt
    OH[i+1] = OH[i]+OH_der[i]*dt
    CVOH[i+1] = CVOH[i]+CVOH_der[i]*dt
    t[i+1] = t[i] + dt
```

```python
#PLOT
pl.title('Reaksjon for dannelse av CVOH')
pl.legend('Reaksjon for dannelse av CVOH')
pl.xlabel('tid (s)')
pl.ylabel('Konsentrasjon mol/L')
pl.plot(t,CVOH, label = 'CVOH', color = 'dimgrey')
pl.plot(t,CV, label = 'CV', color = 'rebeccapurple')
pl.legend(loc=0)
```
